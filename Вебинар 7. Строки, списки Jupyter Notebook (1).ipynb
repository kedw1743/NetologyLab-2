Модификация списков

В примере ниже переменные a и b на самом деле указывают на один и тот же объект. В результате при добавлении в b очередного элемента этот элемент добавляется и в исходном листе a:

```python
a = [1, 2, 3]
b = a
```

```python
b.append(4)
'a = {}'.format(a)
```

```python
id(a), id(b)
```

```python
import copy
```

```python
a = [1, 2, 3]
b = copy.copy(a)
```

```python
id(a), id(b)
```

```python
b.append(4)
print('a = {}'.format(a))

print('b = {}'.format(b))
```

```python
# List comprehension
```

```python
# Дана последовательность чисел. Мы хотим оставить только те, что делятся на 5
sequence = range(0, 40, 3)
list(sequence)
```

```python
# решение в лоб
for num in sequence:
    if num % 5 == 0:
        print(num)
```

```python
# если хотим получить отфильтрованный лист, то будет даже так
filtered_sequence = []

for num in sequence:
    if num % 5 == 0:
        filtered_sequence.append(num)

print(filtered_sequence)
```

```python
[num for num in sequence if num % 5 == 0]
```

```python
[num for num in sequence if num % 5 == 0]
```

```python
[w for w in sequence if w % 5 == 0]
```

```python
# с list comprehension получается покороче

[x**2 for x in sequence if x % 5 == 0]
```
